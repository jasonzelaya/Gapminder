```python
# Import dependencies
import os
import pandas as pd
import numpy as np
import IPython
import country_converter as coco
cc = coco.CountryConverter()
```

```python
# View files in 'raw_data'
os.listdir("raw_data")
```

```python
# Read data into pd dataframes
income = pd.read_csv("raw_data/income_per_person_gdppercapita_ppp_inflation_adjusted.csv")
lex = pd.read_csv("raw_data/life_expectancy_years.csv")
pop = pd.read_csv("raw_data/population_total.csv")
regions4 = pd.read_csv("raw_data/ddf--entities--geo--world_4region.csv")
```

```python
# Preview dataframes
display(income.head())
display(lex.head())
display(pop.head())
display(regions4.head())
```

```python
# View number of values in 'country' column for each dataframe
display(income.country.count())
display(lex.country.count())
display(pop.country.count())
```

```python
# Get only values present in all 'country' columns as a list
country_matches = list(set(income.country) & set(lex.country) & set(pop.country))

# Confirm number of 'country' column values does not exceed 187
display(len(country_matches))

# View list
display(country_matches)
```

```python
# Filter dataframes by matching 'country' column values
income_matches = income[income['country'].isin(country_matches)]
lex_matches = lex[lex['country'].isin(country_matches)]
pop_matches = pop[pop['country'].isin(country_matches)]

# Confirm all dataframes' 'country' columns have 187 rows
display(income_matches.country.count())
display(lex_matches.country.count())
pop_matches.country.count()
```

```python
# Sort dataframes by 'country' columns
income_sorted_matches = income_matches.sort_values(by=['country'])
lex_sorted_matches = lex_matches.sort_values(by=['country'])
pop_sorted_matches = pop_matches.sort_values(by=['country'])

# Preview dataframes
display(income_sorted_matches.head(3))
display(lex_sorted_matches.head(3))
display(pop_sorted_matches.head(3))
```

```python
# Get sorted 'country' column values as a list
sorted_countries = list(income_sorted_matches.country)

# Preview list
sorted_countries[:5]
```

```python
# Convert 'sorted_countries' values to continents
continent_convert = cc.convert(sorted_countries, to='continent')

# Preview cointinents
continent_convert[:5]
```

```python
# Add continents list to each dataframe
income_sorted_matches["region"] = continent_convert
lex_sorted_matches["region"] = continent_convert
pop_sorted_matches["region"] = continent_convert
```

```python
# Preview each dataframe
display(income_sorted_matches.head(3))
display(lex_sorted_matches.head(3))
display(pop_sorted_matches.head(3))
```

```python
# View unique values in 'region' column
display(income_sorted_matches.region.unique())
display(lex_sorted_matches.region.unique())
display(pop_sorted_matches.region.unique())
```

```python
# To determine what to do with "Oceania" since Gapminder is using a 4 region dataset, match Oceania's
# definition with Gapminder's
pd.set_option('display.max_colwidth', -1)
regions4[["world_4region", "name_long"]]
```

```python
# Wikipedia definitions for Australia, Pacific Islands, and South & South East Asia all describe their
# relationship with Oceania
# 'Oceania' values will be converted to 'Asia' values

# Number of 'Asia' rows before Oceania conversion
display(income_sorted_matches.groupby("region").size())
display(lex_sorted_matches.groupby("region").size())
display(pop_sorted_matches.groupby("region").size())
```

```python
# Convert "Oceania" values to "Asia" values
income_sorted_matches["region"] = np.where((income_sorted_matches["region"] == "Oceania"), "Asia", income_sorted_matches["region"])
lex_sorted_matches["region"] = np.where((lex_sorted_matches["region"] == "Oceania"), "Asia", lex_sorted_matches["region"])
pop_sorted_matches["region"] = np.where((pop_sorted_matches["region"] == "Oceania"), "Asia", pop_sorted_matches["region"])
```

```python
# Confirm changes successful
display(income_sorted_matches.region.unique())
display(lex_sorted_matches.region.unique())
pop_sorted_matches.region.unique()
```

```python
# Number of 'Asia' rows after Oceania conversion
display(income_sorted_matches.groupby("region").size())
display(lex_sorted_matches.groupby("region").size())
display(pop_sorted_matches.groupby("region").size())
```

```python
# Preview dataframes
display(income_sorted_matches.head(3))
display(lex_sorted_matches.head(3))
display(pop_sorted_matches.head(3))
```

```python
# Save dataframes into CSV files
income_sorted_matches.to_csv("income_matches.csv", sep=",", index=False)
lex_sorted_matches.to_csv("lex_matches.csv", sep=",", index=False)
pop_sorted_matches.to_csv("pop_matches.csv", sep=",", index=False)
```

```python
# Read data into pd dataframes
income_pivot = pd.read_csv("income_pivoted.csv")
lex_pivot = pd.read_csv("lex_pivoted.csv")
pop_pivot = pd.read_csv("pop_pivoted.csv")
```

```python
# Preview dataframes
display(income_pivot.head())
display(lex_pivot.head())
display(pop_pivot.head())
```

```python
# Remove unnecessary columns
income_drop = income_pivot.drop(columns=["Number of Records"])
lex_drop = lex_pivot.drop(columns=["Number of Records"])
pop_drop = pop_pivot.drop(columns=["Number of Records"])

# Confirm successful removal
display(income_drop.head(2))
display(lex_drop.head(2))
display(pop_drop.head(2))
```

```python
# View min and max 'year' column values
display(income_drop.year.min())
display(lex_drop.year.min())
display(pop_drop.year.min())

display(income_drop.year.max())
display(lex_drop.year.max())
display(pop_drop.year.max())
```

```python
# Remove all rows with 'year' values greater than '2018'
income_2018 = income_drop[income_drop["year"] > 2018 ].index
pop_2018 = pop_drop[pop_drop["year"] > 2018 ].index

income_drop.drop(income_2018, inplace=True)
pop_drop.drop(pop_2018, inplace=True)
```

```python
# Confirm successful removal
display(income_drop.year.max())
display(lex_drop.year.max())
display(pop_drop.year.max())
```

```python
# Confirm row count still matches
display(income_drop.count())
display(lex_drop.count())
display(pop_drop.count())
```

```python
# Check for row count discrepancies
display(income_drop.isnull().sum())
display(lex_drop.isnull().sum())
display(pop_drop.isnull().sum())
```

```python
# Sort dataframes by 'country' columns
income_sort = income_drop.sort_values(by=['country'])
lex_sort = lex_drop.sort_values(by=['country'])
pop_sort = pop_drop.sort_values(by=['country'])

# Preview dataframes
display(income_sort.head(3))
display(lex_sort.head(3))
display(pop_sort.head(3))
```

```python
# Get column values to add to a final dataframe as a list to maintain order of column values
lex_list = list(lex_sort["life_expectancy"])
pop_list = list(pop_sort["population"])

# Preview lists
display(lex_list[:5])
display(pop_list[:5])
```

```python
# Combine data as new dataframe
all_data = income_sort
all_data["life_expectancy"] = lex_list
all_data["population"] = pop_list

# Preview dataframe
all_data.head()
```

```python
# Confirm total row counts are still 40953
display(all_data.count())
display(all_data.isnull().sum())
```

```python
# Sort dataframe by 'year' column
sorted_data = all_data.sort_values(by=['year'])

# Preview dataframe
sorted_data.head(3)
```

```python
# Reorder columns
reorder_data = sorted_data[["country", "region", "year", "income", "life_expectancy", "population"]]

# Confirm reorder successful
reorder_data.head(3)
```

```python
# Remove rows with null values
clean_data = reorder_data.dropna()

# Preview dataframe
display(clean_data.head())
```

```python
# Confirm null values dropped
display(clean_data.isnull().sum())

# Final row count check
display(clean_data.count())
```
