```python
# Import dependencies
import os
import pandas as pd
import numpy as np
import country_converter as coco
cc = coco.CountryConverter()
```

```python
# View files in 'raw_data'
os.listdir("raw_data")
```

```python
# Read data into pd data frames
# Change encoding from ANSI to UTF-8
income = pd.read_csv("raw_data/income_per_person_gdppercapita_ppp_inflation_adjusted.csv", encoding="utf-8")
lex = pd.read_csv("raw_data/life_expectancy_years.csv", encoding="utf-8")
pop = pd.read_csv("raw_data/population_total.csv", encoding="utf-8")
regions4 = pd.read_csv("raw_data/ddf--entities--geo--world_4region.csv", encoding="utf-8")
```

```python
# Preview data frames
display(income.head())
display(lex.head())
display(pop.head())
display(regions4.head())
```

```python
# View number of values in 'country' column for each data frame
display(income["country"].count())
display(lex["country"].count())
display(pop["country"].count())
```

```python
# Get only values present in all 'country' columns as a list
country_matches = list(set(income["country"]) & set(lex["country"]) & set(pop["country"]))

# Confirm number of 'country' column values does not exceed 187
print(len(country_matches))

# Preview list
print(country_matches[:5])
```

```python
# Filter data frames by matching 'country' column values
income_matches = income[income['country'].isin(country_matches)]
lex_matches = lex[lex['country'].isin(country_matches)]
pop_matches = pop[pop['country'].isin(country_matches)]

# Confirm all data frames' 'country' columns have 187 rows
print(income_matches["country"].count())
print(lex_matches["country"].count())
print(pop_matches["country"].count())
```

```python
# Sort data frames by 'country' columns
income_sorted_matches = income_matches.sort_values(by=['country'])
lex_sorted_matches = lex_matches.sort_values(by=['country'])
pop_sorted_matches = pop_matches.sort_values(by=['country'])

# Preview dataframes
display(income_sorted_matches.head(3))
display(lex_sorted_matches.head(3))
display(pop_sorted_matches.head(3))
```

```python
# Convert 'country' column values to continents and add the values as a new "Region" column in the data frames
income_sorted_matches["region"] = cc.convert(list(income_sorted_matches["country"]), to='continent')
lex_sorted_matches["region"] = cc.convert(list(lex_sorted_matches["country"]), to='continent')
pop_sorted_matches["region"] = cc.convert(list(pop_sorted_matches["country"]), to='continent')
```

```python
# Preview each dataframe
display(income_sorted_matches.head(3))
display(lex_sorted_matches.head(3))
display(pop_sorted_matches.head(3))
```

```python
# Confirm all "Region" values converted successfully
print(income_sorted_matches["region"].unique())
print(lex_sorted_matches["region"].unique())
print(pop_sorted_matches["region"].unique())
```

```python
# To determine what to do with "Oceania" since Gapminder is using a 4 region dataset, match Oceania's
# definition with Gapminder's
pd.set_option('display.max_colwidth', -1)
regions4[["world_4region", "name_long"]]
```

```python
# Wikipedia definitions for Australia, Pacific Islands, and South & South East Asia all describe their
# relationship with Oceania
# 'Oceania' values will be converted to 'Asia' values

# Number of 'Asia' rows before Oceania conversion
display(income_sorted_matches.groupby("region").size())
display(lex_sorted_matches.groupby("region").size())
display(pop_sorted_matches.groupby("region").size())
```

```python
# Convert "Oceania" values to "Asia" values
income_sorted_matches["region"] = np.where((income_sorted_matches["region"] == "Oceania"), "Asia", income_sorted_matches["region"])
lex_sorted_matches["region"] = np.where((lex_sorted_matches["region"] == "Oceania"), "Asia", lex_sorted_matches["region"])
pop_sorted_matches["region"] = np.where((pop_sorted_matches["region"] == "Oceania"), "Asia", pop_sorted_matches["region"])
```

```python
# Confirm changes successful
print(income_sorted_matches["region"].unique())
print(lex_sorted_matches["region"].unique())
print(pop_sorted_matches["region"].unique())
```

```python
# Number of 'Asia' rows after Oceania conversion
display(income_sorted_matches.groupby("region").size())
display(lex_sorted_matches.groupby("region").size())
display(pop_sorted_matches.groupby("region").size())
```

```python
# Preview dataframes
display(income_sorted_matches.head(3))
display(lex_sorted_matches.head(3))
display(pop_sorted_matches.head(3))
```

```python
# Save dataframes into CSV files
income_sorted_matches.to_csv("income_matches.csv", sep=",", index=False)
lex_sorted_matches.to_csv("lex_matches.csv", sep=",", index=False)
pop_sorted_matches.to_csv("pop_matches.csv", sep=",", index=False)
```

```python
# Read data into pd data frames
income_pivot = pd.read_csv("income_pivoted.csv")
lex_pivot = pd.read_csv("lex_pivoted.csv")
pop_pivot = pd.read_csv("pop_pivoted.csv")
```

```python
# Preview data frames
display(income_pivot.head())
display(lex_pivot.head())
display(pop_pivot.head())
```

```python
# Remove unnecessary columns
income_drop = income_pivot.drop(columns=["Number of Records"])
lex_drop = lex_pivot.drop(columns=["Number of Records"])
pop_drop = pop_pivot.drop(columns=["Number of Records"])

# Confirm successful removal
print(income_drop.columns)
print(lex_drop.columns)
print(pop_drop.columns)
```

```python
# View min and max 'year' column values
print(income_drop["year"].min())
print(lex_drop["year"].min())
print(pop_drop["year"].min())

print(f"\n{income_drop['year'].max()}")
print(lex_drop["year"].max())
print(pop_drop["year"].max())
```

```python
# Filter all rows with 'year' values greater than '2018'
income_2018 = income_drop.loc[income_drop["year"] < 2019]
pop_2018 = pop_drop.loc[pop_drop["year"] < 2019]

# Confirm removal successful
print(income_2018["year"].max())
print(pop_2018["year"].max())
```

```python
# Keep data frame variables consistent
lex_2018 = lex_drop.copy()

# Preview data frame
lex_2018.head()
```

```python
# Confirm row count still matches
display(income_2018.count())
display(lex_2018.count())
display(pop_2018.count())
```

```python
# Check for row count discrepancies
display(income_2018.isnull().sum())
display(lex_2018.isnull().sum())
display(pop_2018.isnull().sum())
```

```python
# Merge all three data frames together
two_merged = pd.merge(income_2018, lex_2018)
all_merged = pd.merge(two_merged, pop_2018)

# Display dimensions
print(all_merged.shape)

# Preview merged data frame
all_merged.head()
```

```python
# Sort dataframe by 'year' column
sorted_merged = all_merged.sort_values(by=['year'])

# Confirm the first year is 1800 and the last year is 2018
display(sorted_merged.head(2))
sorted_merged.tail(2)
```

```python
# Reorder columns
reordered_merged = sorted_merged[["year", "country", "region", "income", "life_expectancy", "population"]]

# Confirm reorder successful
reordered_merged.head(3)
```

```python
# Remove rows with null values
clean_merged_data = reordered_merged.dropna()

clean_merged_data.reset_index(drop=True, inplace=True)

# Preview data frame
clean_merged_data.head()
```

```python
# Save data frame into CSV file
clean_merged_data.to_csv("all_data_cleaned.csv", sep=",", index=False)
```
